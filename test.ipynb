```python
items = generate_items_from_schedule('Fall 2023 Schedule-2.xlsx')
#items = generate_items_from_schedule('fall2023schedule.xlsx')

conflicts, constraints = gen_conflict_matrix(items)

def gen_students(num):
    students = []
    for i in range(num):
        student = Student2(items, conflicts, constraints)
        students.append(student)
    return students

test_students = gen_students(100)
```
